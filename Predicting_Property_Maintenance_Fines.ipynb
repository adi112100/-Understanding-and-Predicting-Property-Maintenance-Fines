### HANDLING IMBALANCE DATA

```python
X = train.iloc[:,:].drop(['compliance'], axis=1)
y = train.iloc[:,-6]
```

#### Using Oversampling method to to balance data

```python
from imblearn.combine import SMOTETomek
from imblearn.under_sampling import NearMiss
```

```python
smk = SMOTETomek(random_state=42)
X_res,y_res=smk.fit_sample(X,y)
```

```python
a = y_res.reset_index()
print(len(a[a['compliance']==0]), len(a[a['compliance']==0]))
```

### PREDICTION

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
 clf = RandomForestClassifier(n_estimators = 10, max_depth = 5).fit(X_res, y_res)
```

```python
ans = clf.predict_proba(df1)
```
